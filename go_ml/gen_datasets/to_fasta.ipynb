```python
import pandas as pd
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

ds_paths = ['datasets/csa_annot.csv', 'datasets/elms_dataset.csv', 'datasets/llps_dataset.csv']
ds_names = ['csa', 'elms', 'llps']
for ds_path, ds_name in zip(ds_paths, ds_names):
    df = pd.read_csv(ds_path, sep='\t')
    df = df.dropna(subset=['UniprotID', 'Sequence'])  # Ensure no NaN values in key columns
    seq_id = df['UniprotID'].tolist()
    sequences = df['Sequence'].tolist()
    seq_records = [SeqRecord(Seq(seq), id=id_) for seq, id_ in zip(sequences, seq_id)]
    SeqIO.write(seq_records, f'datasets/{ds_name}.fasta', 'fasta')
```
